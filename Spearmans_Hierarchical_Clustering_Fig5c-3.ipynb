#### 2. Compute the 'stretching' of the bigrams, and assign proportion of edits at each genomic position (0-60bp)

```python
import sys
import os
import re
import numpy as np
import collections
from collections import defaultdict
import os.path
import pandas as pd
import math
from math import gcd
import csv

### USER INPUTS: ###

signatures = {'C-odd': 'C',  
              'c-even': 'C', 
              'L-odd': 'I', 
              'l-even': 'I',
              'T-odd': 'D', 
              't-even': 'D'}        #what are the signatures for propagating partners (i.e., 1 A->B and 1 B->A pegRNA controlled by the same inducer), and what signal do they correspond to?
short_list_signals = ['C', 'I', 'D'] #give a list of the signals that are being recorded (1-letter notation)
lists_of_lengths_to_tally = [[2,3,4,5]] #what length records (how many insertions) do you want to be stretched + summed together?  Multiple lengths within one list will be summed together.
values_for_lcm_calc = [2,3,4,5] #what set of numbers should be used to calculate the LCM for the "stretch factor"?

input_identifiers = []     #what sample numbers are you analyzing?  
for num in range(27,81):
    input_identifiers.append(num)
    
    

### END OF USER INPUTS ###



### READING IN EXPERIMENTAL DATA ###

for file_num in input_identifiers:
    #length_counter = 1 # this counter will instruct the appropriate name for the output file.  It describes which lengths were being stretched + summed together.
    for lengths_to_tally in lists_of_lengths_to_tally:
        F_sample_file = f"files/PER71_sig_dataframes_expected/PER71-{file_num}_signatures_dataframe_all_expected_sigs.csv" #input file name here (fastq format; we analyzed fastq files that had F and R reads already assembled via PEAR)
        raw_data = pd.read_csv(F_sample_file)
        data = raw_data.dropna(axis=0) #remove rows that have NaN (it's just one row, because a wt record is denoted as an empty string).  This avoids errors, plus we don't want to stretch the wt records.
        columns = data.columns.tolist()
        columns[0] = "signature"
        data.columns = columns

        edit_labels = "signature"
        list_of_edit_types = []
        for row in data.loc[:, edit_labels]: #make a list of the descriptions of edit types
            list_of_edit_types.append(row)

        epoch_combo = 'proportion' #this is just the header in the input file 
        list_of_edit_proportions = []
        correct_header = '{epochs}'.format(epochs = epoch_combo) #this specificies which column in the input csv file corresponds to the epoch combo you want to plot
        for row in data.loc[:, correct_header]:
            list_of_edit_proportions.append(row)


        ### CONVERTING RAW RECORDS TO 'sig-odd sig-even' format, then converting the signatures to their corresponding signals/inducers ###

        list_of_odd_v_even_edit_types = [] 
        for record in list_of_edit_types:
            counter = 1
            converted_record = []
            for ins in record:
                if counter%2 == 1:
                    converted_ins = ins + '-odd'
                else:
                    converted_ins = ins + '-even'
                converted_record.append(converted_ins)
                counter += 1
            list_of_odd_v_even_edit_types.append(converted_record)

        list_of_edits_wrt_signals = []
        for record in list_of_odd_v_even_edit_types: #converting 'sig-odd' and 'sig-even' to the one-letter symbol corresponding to the appropriate signal
            converted_record = ''
            for ins in record:
                #print(ins)
                if ins in signatures.keys():
                    old_ins = ins
                    ins = signatures[old_ins]
                else:
                    ins = '.' #if the insertion doesn't correspond to an expected signature, replace it with a character designated for unknown signatures
                converted_record += ins
            list_of_edits_wrt_signals.append(converted_record)
                    

        ### "STRETCHING OUT" RECORDS AND ADDING THEM TOGETHER TO DO SIGNAL RECONSTRUCTION ###

        compare = 1 #this is a starting point to find the longest record length that should be used for the stretch factor calculation
        for val in values_for_lcm_calc:
                if val > compare:
                    compare = val

        longest_length = compare

        list_of_lengths = []
        for num in range(1, longest_length +1):
            list_of_lengths.append(num)

        lcm = 1
        for i in list_of_lengths:
            lcm = lcm*i // gcd(lcm,i) #calculate the least common multiple of the record lengths.  All records will be stretched to this length.

        x_axis = lcm

        df_length_frequencies = []
        list_length_frequencies = [] #for each individual read length, add up the frequency of all records of that length.  This will be used for weighting.
        for num in lengths_to_tally:
            list_length_frequencies.append(num)
            frequency_of_length_L = 0
            for record,frequency in zip(list_of_edit_types, list_of_edit_proportions):
                if len(record) == num:
                    frequency_of_length_L += frequency
            list_length_frequencies.append(frequency_of_length_L)
            df_length_frequencies.append(list_length_frequencies)
            list_length_frequencies = []

        max_frequency = 0
        for length in df_length_frequencies: #figure out the read length with the highest frequency, to determine how to weight the other read lengths
            if length[1] > max_frequency:
                max_frequency = length[1]

        if max_frequency == 0: #if there are no records corresponding to the length of interest, skip this length and move on to the next
            length_counter += 1
            continue

        list_stretched_recs = []
        for record, frequency in zip(list_of_edits_wrt_signals, list_of_edit_proportions): #stretch out each record to a consistent length
            if len(record) in lengths_to_tally: #filter out records that have more or less insertions than we want to consider
                raw_length = len(str(record)) #how many insertions are in the record?
                stretch_factor = int(x_axis / raw_length) 
                
                stretched_rec = ''
                for ins in record:
                    stretched_rec = stretched_rec + (ins*stretch_factor) #stretching the record
                
                weight = 0
                for length in df_length_frequencies: #figure out what weight to assign to the record, considering its length
                    if len(record) == length[0]:
                        if length[1] != 0:
                            weight = max_frequency / length[1]
                        else:
                            weight = 0
                        list_stretched_recs.append([frequency*weight, stretched_rec])
                
            
        record_data = []
        record_collection = []    
        for signal in short_list_signals: #tally up ALL stretched records.  One list is being made for the counts of each signature in each stretched record.  This will be compiled into a pandas dataframe and summed up to get total counts of each signature.
            for record in list_stretched_recs:
                record_data.append(signal)
                for ins in record[1]:
                    if ins == signal:
                        record_data.append(record[0])
                    else:
                        record_data.append(0)
                record_collection.append(record_data)
                record_data = []

        #output_lengths = 'undefined'
        # if length_counter == 1:
        #     output_lengths = '1ins'
        # if length_counter == 2:
        #     output_lengths = '2ins'
        # if length_counter == 3:
        #     output_lengths = '3ins'
        # if length_counter == 4:
        #     output_lengths = '4ins'
        # if length_counter == 5:
        #     output_lengths = '5ins'
        # if length_counter == 6:
        output_lengths = 'all2-5ins'
        # if length_counter == 7:
        #     output_lengths = 'all2-4ins'   
            
        stretched_dataframe = pd.DataFrame(record_collection)
        summed_stretched_dataframe = stretched_dataframe.groupby(by=[0]).sum().to_csv(f"files/PER71_STRETCHED_RECORDS/PER71-{file_num}_STRETCHED_RECORDS_{output_lengths}.csv") #before exporting to Excel, sum up all of the counts of each pair at each position along the x-axis

        #length_counter += 1
```

#### Computing Spearmans Correlation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import spearmanr
from scipy.cluster.hierarchy import linkage, dendrogram
import seaborn as sns

#### DEFINE FUNCTIONS ####

# Function for calculating spearmans correlation between two datasets
def calculate_spearmanr(data1, data2):
    # Calculate Spearman's correlation coefficient
    spearman_coeff, _ = spearmanr(data1, data2) 
    return spearman_coeff # returns a float value of coefficient


# Function for plotting heatmap
def plot_heatmap(spearman_coeff_matrix, heatmap_title):
    
    # Set font to Arial
    plt.rcParams['font.family'] = 'Arial'
    
    # Plot the (clustered) heatmap
    sns.set(font_scale = 1.0)
    plt.figure(figsize=(10, 8))
    ax = sns.heatmap(spearman_coeff_matrix, cmap='viridis', annot=True, fmt=".2f", annot_kws={"size": 6.5})
    
    # Set custom tick labels (signal) at the center of grid cells:
    # For X axis
    ax.set_xticks(np.arange(27) + 0.5)
    ax.set_xticklabels(['DDD', 'DDI', 'DD-', 'DID', 'DII', 'DI-', 'D-D', 'D-I', 'D--', 'IDD', 'IDI', 'ID-', 'IID', 'III', 'II-', 'I-D', 'I-I', 'I--', '-DD', '-DI', '-D-', '-ID', '-II', '-I-', '--D', '--I', '---'], rotation=90, ha="center", fontsize= 11)
    # For Y axis:
    ax.set_yticks(np.arange(27) + 0.5)              
    ax.set_yticklabels(['DDD', 'DDI', 'DD-', 'DID', 'DII', 'DI-', 'D-D', 'D-I', 'D--', 'IDD', 'IDI', 'ID-', 'IID', 'III', 'II-', 'I-D', 'I-I', 'I--', '-DD', '-DI', '-D-', '-ID', '-II', '-I-', '--D', '--I', '---'], rotation=0, fontsize= 11)

    # Set axis labels and title
    ax.set_xlabel("Biological Replicate 2 Samples", fontsize= 15)
    ax.set_ylabel("Biological Replicate 1 Samples", fontsize= 15)
    ax.set_title(heatmap_title, fontsize= 18, y=1.02)
    
    # Add outline around heatmap
    for spine in plt.gca().spines.values():
        spine.set_visible(True)
        spine.set_color('black')
        spine.set_linewidth(1)
    
     # Save the figure with 1200 dpi
    #plt.savefig("files/PER71_Spearman_Coefficient_Matrices/PER71-avg_heatmap.png", dpi=1200)

    # Display the heatmap
    plt.show()


# Function for performing hierarchial clustering on matrix of spearman coefficients
def cluster_matrix(spearman_coeff_matrix):
    
    # Define the original order of x/y-axis labels (signals)
    original_labels = ['DDD', 'DDI', 'DD-', 'DID', 'DII', 'DI-', 'D-D', 'D-I', 'D--', 'IDD', 'IDI', 'ID-', 'IID', 'III', 'II-', 'I-D', 'I-I', 'I--', '-DD', '-DI', '-D-', '-ID', '-II', '-I-', '--D', '--I', '---']

    # Perform hierarchical clustering on spearman_coeff_matrix
    Z = linkage(spearman_coeff_matrix, method='single', metric='euclidean')

    # Obtain the leaf order from the hierarchical clustering results
    leaf_order = dendrogram(Z, no_plot=True)['leaves']

    # Reorder the original labels based on the leaf order
    clustered_matrix_labels = [original_labels[i] for i in leaf_order]

    # Reorder the rows and columns based on the leaf order
    clustered_matrix = spearman_coeff_matrix[leaf_order, :][:, leaf_order]

    return clustered_matrix, clustered_matrix_labels


# Function for plotting clustered matrix
def plot_clustered_heatmap(clustered_matrix, clustered_matrix_labels, clustered_matrix_title):
    
    # Set font to Arial
    plt.rcParams['font.family'] = 'Arial'
    
    # Plot the (clustered) heatmap
    sns.set(font_scale = 1.0)
    plt.figure(figsize=(10, 8))
    ax = sns.heatmap(clustered_matrix, cmap='viridis', annot=True, fmt=".2f", annot_kws={"size": 6.5})
    
    # Set x- and y-axis tick and labels
    ax.set_xticks(np.arange(len(clustered_matrix_labels)) + 0.5)
    ax.set_xticklabels(clustered_matrix_labels, rotation=90, ha="center", fontsize= 11)
    
    ax.set_yticklabels(clustered_matrix_labels, rotation=0, fontsize= 11)

    ax.set_xlabel("Biological Replicate 2 Samples", fontsize= 15)
    ax.set_ylabel("Biological Replicate 1 Samples", fontsize= 15)
    ax.set_title(clustered_matrix_title, fontsize= 18, y=1.02)
    
    # Add outline around heatmap
    for spine in plt.gca().spines.values():
        spine.set_visible(True)
        spine.set_color('black')
        spine.set_linewidth(1)
        
    # Save the figure with 1200 dpi
    #plt.savefig("files/PER71_Spearman_Coefficient_Matrices/PER71-clustered_avg_heatmap.png", dpi=1200)

    # Display the clustered heatmap
    plt.show()    
    


#### MAIN PROGRAM ####

# Make sub-lists of samples in groups/replicates 1 and 2 (to compare)
sub_list1 = []     
for num in range(27,54):
    sub_list1.append(num)

sub_list2 = []     
for num in range(54,81):
    sub_list2.append(num)
    
# Initialize lists to store values of spearman coefficients when comparing dox vs iptg proportions
spearman_coefficients_D = []
spearman_coefficients_I = []
spearman_coefficients_avg = [] # average of the above two coefficients


# Iterate over every replicate pair and compute spearman coefficients
for sample1 in sub_list1:
    spearman_co_D = []
    spearman_co_I = []
 
    file1 = "files/PER71-D_I_separate_lines/PER71-{}_D_I_separate_lines.csv".format(sample1)
    df5 = pd.read_csv(file1, header=None, skiprows=1)
    sample_list1D = df5[0].to_list() #create a list of Dox proportions
    sample_list1I = df5[1].to_list() #create a list of IPTG proportions

    for sample2 in sub_list2:
        file2 = "files/PER71-D_I_separate_lines/PER71-{}_D_I_separate_lines.csv".format(sample2)
        df6 = pd.read_csv(file2, header=None, skiprows=1)
        sample_list2D = df6[0].to_list() # Dox proportions list
        sample_list2I = df6[1].to_list() # IPTG proportions list
        
        # Calculate spearman coefficients (for each of Dox and IPTG proportions) between replicate pair
        spearman_D = calculate_spearmanr(sample_list1D, sample_list2D)
        spearman_co_D.append(spearman_D)
        spearman_I = calculate_spearmanr(sample_list1I, sample_list2I)
        spearman_co_I.append(spearman_I)
        
        # Calculate the average of the spearman coefficients above (Dox, IPTG)
        spearman_co_avg = []
        for i in range(len(spearman_co_D)):
            spearman_avg = (spearman_co_D[i] + spearman_co_I[i]) / 2
            spearman_co_avg.append(spearman_avg)
    
    
    spearman_coefficients_D.append(spearman_co_D)
    spearman_coefficients_I.append(spearman_co_I)
    spearman_coefficients_avg.append(spearman_co_avg) 
            
        
# Convert spearman_coefficients to a NumPy array
spearman_coefficients_D_array = np.array(spearman_coefficients_D)
spearman_coefficients_I_array = np.array(spearman_coefficients_I)
spearman_coefficients_avg_array = np.array(spearman_coefficients_avg)


# Plot heatmaps of spearman coefficient matrices
#plot_heatmap(spearman_coefficients_D_array, 'Spearman coefficients of Dox proportions between replicates')
#plot_heatmap(spearman_coefficients_I_array, 'Spearman coefficients of IPTG proportions between replicates')
plot_heatmap(spearman_coefficients_avg_array, 'Spearman coefficients of avg(D,I) proportions between replicates')


# Perform clustering on spearman coefficient matrices
clustered_avg_matrix, clustered_avg_matrix_labels = cluster_matrix(spearman_coefficients_avg_array)
# can also perform clustering on Dox and IPTG matrices if desired

# Plot heatmaps of clustered matrices
plot_clustered_heatmap(clustered_avg_matrix, clustered_avg_matrix_labels, "Hierarchial clustered matrix of Spearman coefficients of avg(D,I)")


## SAVING FILES 

# Saving (original matrix) heatmaps to a csv file
outfile_spear_coeff_D = "files/PER71_Spearman_Coefficient_Matrices/SpearCoeff_Matrix_D.csv"
outfile_spear_coeff_I = "files/PER71_Spearman_Coefficient_Matrices/SpearCoeff_Matrix_I.csv"
outfile_spear_coeff_DI_avg = "files/PER71_Spearman_Coefficient_Matrices/SpearCoeff_Matrix_DIavg.csv"

np.savetxt(outfile_spear_coeff_D,spearman_coefficients_D_array,delimiter=',')
np.savetxt(outfile_spear_coeff_I,spearman_coefficients_I_array,delimiter=',')
np.savetxt(outfile_spear_coeff_DI_avg,spearman_coefficients_avg_array,delimiter=',')


# Saving (clustered) heatmaps to a csv file
outfile_spear_coeff_DI_avg_clustered = "files/PER71_Spearman_Coefficient_Matrices/SpearCoeff_Matrix_DIavg_Clustered.csv"
np.savetxt(outfile_spear_coeff_DI_avg_clustered,clustered_avg_matrix,delimiter=',')
```
